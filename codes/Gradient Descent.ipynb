# Plain Gradient Descent 

```python
def LearningSchedule(t):
    """
    Computes the learning rate based on the iteration number.
    
    Parameters:
    - t (int): Current iteration number.
    
    Returns:
    - float: Adjusted learning rate for the current iteration.
    """
    return t0 / (t + t1)


def GradientDescent(X, y, Beta, eta, Niterations, rho, lmbda=None):
    """
    Performs gradient descent optimization with optional momentum and ridge regularization.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - Beta (array): Initial weights/coefficients.
    - eta (float): Learning rate.
    - Niterations (int): Number of iterations for the gradient descent loop.
    - rho (float): Momentum parameter (0 for no momentum).
    - lmbda (float, optional): Ridge regularization parameter (default: None).

    Returns:
    - MSE (list): Mean Squared Error for each iteration.
    - Beta (array): Updated weights/coefficients after training.
    """

    # Initializing list to store MSE values for each iteration
    MSE = []

    # Initializing velocity for momentum updates, if rho > 0
    velocity = np.zeros_like(Beta)
    old_velocity = 0  # Keeps track of previous velocity for momentum

    for iteration in range(Niterations):
        # Computing gradient of the cost function
        gradient = (2.0 / n) * X.T @ (X @ Beta - y)
        
        # Adding Ridge regularization if lambda is specified
        if lmbda is not None:
            gradient += 2 * lmbda * Beta
        
        # Updating parameters with or without momentum
        if rho > 0:
            # Updating with momentum
            velocity = rho * old_velocity + eta * gradient
            Beta -= velocity
            old_velocity = velocity  # Update old velocity for the next iteration
        else:
            # Standard gradient descent update without momentum
            Beta -= eta * gradient

        # Calculating and store the MSE for the current iteration
        y_pred = X @ Beta
        MSE.append(mean_squared_error(y, y_pred))

    return MSE, Beta
```

# Stochastic Gradient Descent

```python
def StochasticGradient(X, y, Beta, M, epochs, rho, lmbda=None, eta_max=1.0, gradient_clip=10):
    """
    Performs stochastic gradient descent with optional momentum and regularization.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - Beta (array): Initial weights/coefficients.
    - M (int): Batch size for mini-batch gradient descent.
    - epochs (int): Number of epochs (complete passes over the dataset).
    - rho (float): Momentum parameter (0 for no momentum).
    - lmbda (float, optional): Regularization parameter (L2 regularization).
    - eta_max (float): Maximum learning rate.
    - gradient_clip (float): Gradient clipping threshold to avoid explosion.

    Returns:
    - Beta (array): Updated weights/coefficients after training.
    - MSE_epochs_ (list): MSE for each epoch.
    - MSE_batch_ (list): MSE for each batch within each epoch.
    - Eta (list): Learning rate (eta) values used for each batch.
    """

    # Tracking MSE for each epoch and batch
    MSE_epochs_ = []
    MSE_batch_ = []
    Eta = []

    # Initializing velocity for momentum, if rho > 0
    velocity = np.zeros_like(Beta)
    old_velocity = 0

    for epoch in range(1, epochs + 1):
        # Shuffling data at each epoch for stochasticity
        indice = np.random.permutation(len(y))
        X_shuffled = X[indice]
        y_shuffled = y[indice]
        MSE_batch_temp = []
        Eta_temp = []

        # Iterating over mini-batches
        for i in range(0, len(y), M):
            # Extracting mini-batch of M samples
            Xi = X_shuffled[i:i + M]
            yi = y_shuffled[i:i + M]
            
            if len(yi) < M:
                continue  # Skipping incomplete batches

            # Calculating gradient for mini-batch
            gradient = (2.0 / len(Xi)) * Xi.T @ ((Xi @ Beta) - yi)
            eta = LearningSchedule(epochs * M + i)
            eta = min(eta, eta_max)  # Cap eta to avoid excessive learning rates
            Eta_temp.append(eta)

            # Applying L2 regularization if lambda is provided
            if lmbda is not None:    
                gradient += 2 * lmbda * Beta

            # Clipping gradient to avoid explosion
            gradient = np.clip(gradient, -gradient_clip, gradient_clip)

            # Updating Beta with momentum if rho > 0
            if rho > 0:
                velocity = rho * old_velocity + eta * gradient
                Beta -= velocity
                old_velocity = velocity
            else:
                # Updating Beta without momentum
                Beta -= eta * gradient

            # Calculating and storing MSE for the batch
            ypred_batch = Xi @ Beta
            MSE_batch_temp.append(mean_squared_error(yi, ypred_batch))

        # Calculating and storing MSE for the epoch
        y_pred_epoch = X @ Beta
        MSE_epochs_.append(mean_squared_error(y, y_pred_epoch))
        MSE_batch_.append(MSE_batch_temp)
        Eta.append(Eta_temp)

    return Beta, MSE_epochs_, MSE_batch_, Eta
```

# Adagrad Plain Gradient Descent

```python
def Adagrad_plain(X, y, Beta, epochs, rho, epsilon):
    """
    Implements the Adagrad optimization algorithm with optional momentum.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - Beta (array): Initial weights/coefficients.
    - epochs (int): Number of training epochs.
    - rho (float): Momentum parameter (0 for no momentum).
    - epsilon (float): Smoothing term to avoid division by zero.

    Returns:
    - MSE (list): Mean Squared Error for each epoch.
    - Beta (array): Updated weights/coefficients after training.
    """
    MSE = []  # Track MSE for each epoch
    
    # Initialize accumulator for squared gradients (Adagrad-specific)
    G = np.zeros_like(Beta)
    velocity = np.zeros_like(Beta)      # Initialize velocity for momentum if rho > 0
    velocity_old = np.zeros_like(Beta)  # Needed for momentum updates in Adagrad

    for epoch in range(1, epochs + 1):
        # Compute gradient of the cost function
        gradient = 2.0 / len(y) * X.T @ (X @ Beta - y)
        
        # Update learning rate using the learning schedule function (could vary with epoch)
        eta = LearningSchedule(epoch)
        
        # Apply Adagrad update with or without momentum
        if rho > 0:
            # Accumulate squared gradients for Adagrad scaling
            G += gradient**2
            # Update velocity with momentum and scale by accumulated gradients
            velocity = rho * velocity_old + eta * gradient / (np.sqrt(G) + epsilon)
            # Update parameters (Beta) using the adjusted gradient
            Beta -= velocity
            velocity_old = velocity  # Update velocity for next iteration
        else:
            # Standard Adagrad without momentum
            G += gradient**2
            Beta -= eta * gradient / (np.sqrt(G) + epsilon)

        # Calculate MSE for the current epoch and store it
        ypred = X @ Beta
        if not np.isnan(ypred).any() and ypred.shape == y.shape:
            MSE.append(mean_squared_error(y, ypred))

    return MSE, Beta

# Parameters for the Adagrad algorithm
t0, t1 = 5, 10  # Parameters for learning rate scheduling (if used in LearningSchedule)
epochs = 100    # Number of epochs for training
rho = 0.01      # Momentum parameter for Adagrad with momentum
Beta = np.random.randn(3, 1)  # Random initialization of Beta
epsilon = 1e-3  # Small value to prevent division by zero in Adagrad

# Copy Beta to run Adagrad with and without momentum consistently
Beta_plain = Beta.copy()
Beta_momentum = Beta.copy()

# Run Adagrad without momentum
adagrad_MSE, adagrad_coef = Adagrad_plain(scaled_X, y, Beta_plain, epochs, 0, epsilon)

# Run Adagrad with momentum
adagrad_momentum_MSE, adagrad_momentum_coef = Adagrad_plain(scaled_X, y, Beta_momentum, epochs, rho, epsilon)

# Plot the Mean Squared Error (MSE) over epochs for both methods
plt.figure(figsize=(12, 6))
plt.plot(range(1, epochs + 1), adagrad_MSE, label='MSE Adagrad')
plt.plot(range(1, epochs + 1),adagrad_momentum_MSE, label='MSE Adagrad with Momentum')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('MSE of Adagrad Optimization with and without Momentum')
plt.grid()
plt.legend()
plt.show()

```

# Adagrad Stochastic Gradient Descent

```python
def Adagrad_StochasticGradient(X, y, Beta, M, epochs, rho, eta, epsilon):
    """
    Implements the Adagrad optimization algorithm with mini-batch SGD and optional momentum.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - Beta (array): Initial weights/coefficients.
    - M (int): Mini-batch size.
    - epochs (int): Number of epochs for training.
    - rho (float): Momentum parameter (0 for no momentum).
    - eta (float): Initial learning rate.
    - epsilon (float): Smoothing term to avoid division by zero.

    Returns:
    - MSE (list): Mean Squared Error for each epoch.
    - Beta (array): Updated weights/coefficients after training.
    """
    MSE = []    # Track MSE over epochs
    Eta_ = []   # Placeholder for recording learning rates if needed
    G = np.zeros_like(Beta)  # Accumulator for squared gradients (Adagrad-specific)
    velocity = np.zeros_like(Beta)      # Initialize velocity for momentum if rho > 0
    velocity_old = np.zeros_like(Beta)  # Velocity tracker for previous iteration

    for epoch in range(1, epochs + 1):
        # Shuffling the data at each epoch for stochastic updates
        shuffled_indices = np.random.permutation(len(y))
        X_shuffled = X[shuffled_indices]
        y_shuffled = y[shuffled_indices]

        # Processing mini-batches
        for i in range(0, len(y), M):
            # Select mini-batch
            Xi = X_shuffled[i:i + M]
            yi = y_shuffled[i:i + M]

            # Ensuring valid batch size to avoid partial mini-batches
            if len(yi) < M:
                continue
            
            # Checking for NaN values in mini-batch (could result from pre-processing issues)
            if np.isnan(Xi).any() or np.isnan(yi).any():
                continue

            # Computing gradient for mini-batch
            gradient = 2.0 / len(y) * Xi.T @ (Xi @ Beta - yi)

            # Applying Adagrad with or without momentum based on rho
            if rho > 0:
                # With Adagrad and momentum
                G += gradient**2
                velocity = rho * velocity_old + eta * gradient / (np.sqrt(G) + epsilon)
                Beta -= velocity
                velocity_old = velocity  # Update velocity for next iteration
            else:
                # Adagrad without momentum
                G += gradient**2
                Beta -= eta * gradient / (np.sqrt(G) + epsilon)
        
        # Calculating and store MSE for the epoch
        ypred = X @ Beta
        if not np.isnan(ypred).any() and ypred.shape == y.shape:
            MSE.append(mean_squared_error(y, ypred))

    return MSE, Beta

# Preparing the design matrix and scale features
X = np.c_[np.ones((n, 1)), x, x * x]  # Design matrix with bias and polynomial features
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

# Initializing parameters for Adagrad
Beta = np.random.randn(3, 1)
M = 10       # Mini-batch size
epochs = 100  # Number of training epochs
rho = 0.1    # Momentum parameter
epsilon = 1e-3  # Small value to prevent division by zero
eta = 0.2    # Learning rate

# Making copies of Beta to test Adagrad with and without momentum
Beta_no_momentum = Beta.copy()
Beta_momentum = Beta.copy()

# Printing initial Beta values to check consistency across runs
print("Initial Beta (no momentum):", Beta_no_momentum)
print("Initial Beta (with momentum):", Beta_momentum)

# Running Adagrad with and without momentum
adagrad_stochastic_MSE, adagrad_coef = Adagrad_StochasticGradient(
    scaled_X, y, Beta_no_momentum, M, epochs, 0, eta, epsilon=epsilon
)
adagrad_stochastic_momentum_MSE, adagrad_momentum_coef = Adagrad_StochasticGradient(
    scaled_X, y, Beta_momentum, M, epochs, rho, eta, epsilon=epsilon
)

# Plotting the Mean Squared Error (MSE) over epochs
plt.figure(figsize=(12, 6))
plt.plot(range(1, epochs + 1), adagrad_stochastic_MSE, label='MSE Adagrad')
plt.plot(range(1, epochs + 1), adagrad_stochastic_momentum_MSE, label='MSE Adagrad with Momentum')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('MSE for Adagrad-SGD with/without Momentum')
plt.grid()
plt.legend()
plt.show()
```

# RMSprop

```python
#Adding RMSprop to the library

def rmsprop(X, y, beta, eta, epochs, rho=0.9, epsilon=1e-8):
    """
    Implements the RMSprop optimization algorithm using full-batch gradient descent.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - beta (array): Initial weights/coefficients.
    - eta (float): Learning rate.
    - epochs (int): Number of epochs for training.
    - rho (float): Decay rate for running average of squared gradients (default: 0.9).
    - epsilon (float): Smoothing term to avoid division by zero (default: 1e-8).

    Returns:
    - beta (array): Updated weights/coefficients after training.
    """
    
    G = np.zeros_like(beta)  # Initializing running average for squared gradients

    for epoch in range(epochs):
        # Compute gradient for the entire dataset
        gradient = (2.0 / X.shape[0]) * X.T @ (X @ beta - y)

        # Update the running average of squared gradients
        G = rho * G + (1 - rho) * gradient**2

        # Update parameters using RMSprop update rule
        beta -= (eta / (np.sqrt(G) + epsilon)) * gradient

    return beta

# Parameters for RMSprop optimization
eta = 0.001   # Learning rate for RMSprop
epochs = 100  # Number of epochs
rho = 0.9     # Decay rate for the running average of squared gradients
beta = np.random.randn(3, 1)  # Initialize beta coefficients randomly

# Running RMSprop with full-batch gradient descent
beta_rmsprop = rmsprop(X, y, beta, eta, epochs, rho)
print("Beta after RMSprop:")
print(beta_rmsprop)
```

# ADAM

```python
#Adding adam to the library
def adam(X, y, beta, eta, epochs, beta1=0.9, beta2=0.999, epsilon=1e-8):
    """
    Implements the Adam optimization algorithm.

    Parameters:
    - X (array): Feature matrix.
    - y (array): Target values.
    - beta (array): Initial weights/coefficients.
    - eta (float): Learning rate.
    - epochs (int): Number of epochs for training.
    - beta1 (float): Decay rate for the first moment estimate (default: 0.9).
    - beta2 (float): Decay rate for the second moment estimate (default: 0.999).
    - epsilon (float): Smoothing term to avoid division by zero (default: 1e-8).

    Returns:
    - beta (array): Updated weights/coefficients after training.
    """
    
    m = np.zeros_like(beta)  # Initializing first moment vector (momentum)
    v = np.zeros_like(beta)  # Initializing second moment vector (squared gradient)
    t = 0  # Time step counter for bias correction

    for epoch in range(epochs):
        t += 1  # Incrementing time step
        
        # Computing gradient for current parameters
        gradient = (2.0 / X.shape[0]) * X.T @ (X @ beta - y)
        
        # Updating biased first moment estimate
        m = beta1 * m + (1 - beta1) * gradient
        
        # Updating biased second moment estimate
        v = beta2 * v + (1 - beta2) * gradient**2
        
        # Correcting bias in first moment estimate
        m_hat = m / (1 - beta1**t)
        
        # Correcting bias in second moment estimate
        v_hat = v / (1 - beta2**t)
        
        # Updating parameters
        beta -= eta * m_hat / (np.sqrt(v_hat) + epsilon)

    return beta

# Parameters for Adam optimization
eta = 0.001   # Learning rate for Adam
epochs = 100  # Number of epochs
beta1 = 0.9   # Decay rate for the first moment estimate
beta2 = 0.999 # Decay rate for the second moment estimate
beta = np.random.randn(3, 1)  # Initialize beta coefficients randomly

# Run Adam optimization
beta_adam = adam(X, y, beta, eta, epochs, beta1, beta2)
print("Beta after Adam:")
print(beta_adam)
```
